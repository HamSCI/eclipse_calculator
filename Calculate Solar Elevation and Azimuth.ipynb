# Solar Elevation and Azimuth vs. Time

```python
import datetime
import matplotlib as mpl
from matplotlib import pyplot as plt
import eclipse_calc

%matplotlib inline
```

```python
plt.rcParams['font.size']        = 18
plt.rcParams['font.weight']      = 'bold'
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.grid']        = True
plt.rcParams['axes.xmargin']     = 0
plt.rcParams['grid.linestyle']   = ':'
```

## Do the Calculations

```python
# Set Lat/Lon to W2NAF in Spring Brook Township, PA
lat =  41.335116
lon = -75.600692
```

```python
# Create a vector of times to evaluate
sTime = datetime.datetime(2024,4,8)
eTime = datetime.datetime(2024,4,9)
dt    = datetime.timedelta(minutes=5)

times = [sTime]
while times[-1] < eTime:
    times.append(times[-1]+dt)
```

```python
# Calculate Solar Azimuth and Elevation Values
az, el  = eclipse_calc.calculate_solarAzEl(times,lat,lon)
```

## Plot the Results

```python
# Plot results - Elevation Angle
fig      = plt.figure(figsize=(14,5))
ax       = fig.add_subplot(111)
ax.plot(times,el)
ax.set_xlabel('Time [UTC]')
ax.set_ylabel('Solar Elevation [deg]')
ax.set_title('{!s}\N{DEGREE SIGN}N, {!s}\N{DEGREE SIGN}E'.format(lat,lon))

# Empahsize the Horizontal Line at 0 deg (horizon)
ax.axhline(0,color='k',ls='--')

# Format X-Axis Times to HHMM for easier readability.
xtks    = ax.get_xticks()
xtk_lbs = []
for xtk in xtks:
    xtk_dt = mpl.dates.num2date(xtk)
    xtl    = xtk_dt.strftime('%H%M')
    xtk_lbs.append(xtl)
ax.set_xticks(xtks)
ax.set_xticklabels(xtk_lbs)

plt.show()
```

```python
# Plot results - Azimuth Angle
fig      = plt.figure(figsize=(14,5))
ax       = fig.add_subplot(111)
ax.plot(times,az)
ax.set_xlabel('Time [UTC]')
ax.set_ylabel('Solar Azimuth [deg]')
ax.set_title('{!s}\N{DEGREE SIGN}N, {!s}\N{DEGREE SIGN}E'.format(lat,lon))

# Format X-Axis Times to HHMM for easier readability.
xtks    = ax.get_xticks()
xtk_lbs = []
for xtk in xtks:
    xtk_dt = mpl.dates.num2date(xtk)
    xtl    = xtk_dt.strftime('%H%M')
    xtk_lbs.append(xtl)
ax.set_xticks(xtks)
ax.set_xticklabels(xtk_lbs)

plt.show()
```
